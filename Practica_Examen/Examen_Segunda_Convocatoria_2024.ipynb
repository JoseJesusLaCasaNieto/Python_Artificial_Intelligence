# 1 - 1.25 pt
Dado una lista de valores, cree un diccionario cuya clave sea el dato de la lista y el valor sea el resto de la división entre 6. No se deben incluir en el diccionario los valores cuyo resto es igual a cero.

```python
# Your solution HERE !!!!
def nombre_modulo(lst):
    return {number: number%6 for number in lst if number%6 != 0}
```

```python
ls_val = [30, 23, 7, 39, 55,  5, -4]
dc_res = nombre_modulo(ls_val)  # This calls your function
dc_solucion = {23: 5, 7: 1, 39: 3, 55: 1, 5: 5, -4: 2}
assert(dc_res == dc_solucion), "Solución incorrecta"
```

# 2 - 1.25 pt

Realice con numpy:
- Una matriz 4x2 (A) y un vector 4x1 (b) con valores aleatorios
- Calcule la inversa de la multiplicación A traspuesta por la misma matriz, multiplíquelo por la matriz traspuesta de A y el vector b: $x =(A^{T}  A)^{-1} A^{T}b$ 
- Multiplique $x$ por un array de dos valores aleatorios.

```python
# Your solution HERE !!!!
A = np.random.randint(0, 21, size=(4, 2))
b = np.random.randint(0, 21, size=(4, 1))
print("A =", A)
print("\nb =", b)

x = np.dot(np.linalg.inv(np.dot(np.transpose(A), A)), np.dot(np.transpose(A), b))
print("\nx =", x)

arr = np.random.randint(0, 21, size=(1, 2))
result = np.dot(x, arr)
print("\n", result)
```

# 3 - 1.25 pt

Se pide crear un generador que recorra una lista dada y, que en cada llamada, devuelva la media de los valores contiguos y el valor central definido por la posición del índice de dicha lista. Es decir, dada la lista [12, 3, 9, 3, 6], el generador devolverá en la primera iteración la media de 12 y 3 dando como resultado 7.5. En la segunda iteración será 12, 3 y 9, resultando 8. El último valor que devolverá será 3 y 6, resultando 4.5.

```python
# Your solution HERE !!!!
def generador_medias(lst):
    position = 0
    while position < len(lst):
        if position == 0:
            yield np.round((lst[position] + lst[position + 1]) / 2, 1)
        elif position == len(lst) - 1:
            yield np.round((lst[position - 1] + lst[position]) / 2, 1)
        else:
            yield np.round((lst[position - 1] + lst[position] + lst[position + 1]) / 3, 1)
        position += 1
```

```python
ls_num = [12, 3, 9, 3, 6]
gen = generador_medias(ls_num)
ls_res = []
ls_solucion = [7.5, 8.0, 5.0, 6.0, 4.5]
for i in gen:
    ls_res.append(i)
assert (ls_res==ls_solucion), "Solución incorrecta"
```

# 4 - 1.25 pt
Se pide crear una función que reciba un array de NumPy y devuelva otro en el cual estén marcados como True los elementos duplicados (a partir de la segunda ocurrencia), y como False los no repetidos o las primeras ocurrencias de los duplicados.

```python
# Your solution HERE !!!!
def find_duplicates(lst):
    result = []
    duplicates = []
    for item in lst:
        if item not in duplicates:
            duplicates.append(item)
            result.append(False)
        else:
            result.append(True)
    return np.array(result)
```

```python
numbers = np.array([4, 4, 4, 3, 8, 1, 9, 6, 1, 5, 9, 4, 0, 2])
expected_output = np.array([False, True, True, False, False, False, False, False, True, False, True, True, False, False])
output = find_duplicates(numbers) # This calls your function find_duplicates
assert(np.array_equal(expected_output, output))  # This will fail if the result is not as expected
```

```python
# Apartado 5
mask = data['Provincias'] == 'Total Nacional'
data4 = data[mask]
mask = data4['Sexo'] == 'Ambos sexos'
data4 = data4[mask]

mask = data4['Tasas'] == 'Tasa de actividad'
data4_actividad = data4[mask]
mask = data4['Tasas'] == 'Tasa de paro de la población'
data4_paro = data4[mask]
mask = data4['Tasas'] == 'Tasa de empleo de la población'
data4_empleo = data4[mask]

fig, axes = plt.subplots(3, 1, figsize=(7, 15))

data4_actividad['Año'] = data4_actividad['Periodo'].str[:4]
data4_actividad['Trimestre'] = data4_actividad['Periodo'].str[-2:]

data4_actividad_clean = data4_actividad[['Total', 'Año', 'Trimestre']]

for year in data4_actividad_clean['Año'].unique():
    data_year = data4_actividad_clean[data4_actividad_clean['Año'] == str(year)]
    data_year = data_year.groupby('Trimestre').mean(numeric_only=True)
    axes[0].plot(data_year)

axes[0].set_title('Tasa de actividad')
axes[0].set_ylabel('Total')

data4_paro['Año'] = data4_paro['Periodo'].str[:4]
data4_paro['Trimestre'] = data4_paro['Periodo'].str[-2:]

data4_paro_clean = data4_paro[['Total', 'Año', 'Trimestre']]

for year in data4_paro_clean['Año'].unique():
    data_year = data4_paro_clean[data4_paro_clean['Año'] == str(year)]
    data_year = data_year.groupby('Trimestre').mean(numeric_only=True)
    axes[1].plot(data_year)

axes[1].set_title('Tasa de paro de la población')
axes[1].set_ylabel('Total')

data4_empleo['Año'] = data4_empleo['Periodo'].str[:4]
data4_empleo['Trimestre'] = data4_empleo['Periodo'].str[-2:]

data4_empleo_clean = data4_empleo[['Total', 'Año', 'Trimestre']]

for year in data4_empleo_clean['Año'].unique():
    data_year = data4_empleo_clean[data4_empleo_clean['Año'] == str(year)]
    data_year = data_year.groupby('Trimestre').mean(numeric_only=True)
    axes[2].plot(data_year)

axes[2].set_title('Tasa de empleo de la población')
axes[2].set_xlabel('Trimestre')
axes[2].set_ylabel('Total')

plt.show()
```
