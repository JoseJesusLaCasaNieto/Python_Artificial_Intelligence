## 02 - 1.25pt
- Escribir un generador que genere todos los factores de un número. Un factor es un número natural que puede dividir exactamente a otro número.

```python
number = 1000
expected_factors = [1, 2, 4, 5, 8, 10, 20, 25, 40, 50, 100, 125, 200, 250, 500, 1000]
gen = factor_generator(number) # This calls tour function factor_generator
output = [next(gen) for n in range(len(expected_factors))]
assert(expected_factors == output)  # This will fail if the element is not found correctly
```

```python
# Your solution HERE !!!!
number = 1000

def factor_generator(n):
    result = []
    for i in range(1, n+1):
        if n % i == 0:
            result.append(i)
    return result

print(factor_generator(number))
```

```python
gen = factor_generator(number)
expected_factors = [1, 2, 4, 5, 8, 10, 20, 25, 40, 50, 100, 125, 200, 250, 500, 1000]
assert(expected_factors == gen)
```

## 04 - 1.25pt
- 200 personas hacen una review de una película, dándole una puntuación de 0.0 a 5.0 a dicha película
- Crear una lista aleatoria con las posibles puntuaciones, es decir, 200 puntuaciones entre 0.0 y 5.0
- Discretizar dichas puntuaciones en 4 categorias [muy mala, mala, buena, excelente], los límites de cada categoría son a libre elección
- Discretizar dichas puntuaciones en 5 categorías distribuidas con rangos homogéneos (similar cantidad de valores en cada categoría)
- Mostrar las categorías creadas y cuantos elementos hay en cada una de ellas

```python
# Your solution HERE !!!!
random_list = []

for i in range(200):
    random_list.append(np.random.random() * 5)

print(random_list)
```

```python
very_bad, bad, good, excellent = [], [], [], []

for item in random_list:
    if item <= 1.5:
        very_bad.append(item)
    elif item > 1.5 and item <= 2.5:
        bad.append(item)
    elif item > 2.5 and item <= 4:
        good.append(item)
    else:
        excellent.append(item)
```

```python
cat_1, cat_2, cat_3, cat_4, cat_5 = [], [], [], [], []

for item in random_list:
    if item <= 1:
        cat_1.append(item)
    elif item > 1 and item <= 2:
        cat_2.append(item)
    elif item > 2 and item <= 3:
        cat_3.append(item)
    elif item > 3 and item <= 4:
        cat_4.append(item)
    else:
        cat_5.append(item)
```

```python
# Discretización 1
print(very_bad)
print("Elementos: ", len(very_bad))
print(bad)
print("Elementos: ", len(bad))
print(good)
print("Elementos: ", len(good))
print(excellent)
print("Elementos: ", len(excellent))

# Discretización 2
print('\n' + 'Discretización 2' + '\n')
print(cat_1)
print("Elementos: ", len(cat_1))
print(cat_2)
print("Elementos: ", len(cat_2))
print(cat_3)
print("Elementos: ", len(cat_3))
print(cat_4)
print("Elementos: ", len(cat_4))
print(cat_5)
print("Elementos: ", len(cat_5))
```
